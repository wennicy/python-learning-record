```python
# 猜数字游戏（3次机会）
import random
num = random.randint(1,10)

# 第一次猜
guess = int(input("请猜一个 1-10 之间的数字:"))

if guess == num:
    print("恭喜你！第一次就猜中了！")
else:
    if guess > num:
        print("你猜错啦！你的结果大了。")
    else:
        print("你猜错啦！你的结果小了。")
    
    # 第二次猜  
    guess = int(input("请再猜一个 1-10 之间的数字:"))

    if guess == num:
        print("恭喜你！第二次猜中了！")
    else:
        if guess > num:
            print("你猜错啦！你的结果大了。")
        else:
            print("你猜错啦！你的结果小了。")
        
        # 第三次猜  
        guess = int(input("请再猜一个 1-10 之间的数字:"))

        if guess == num:
            print("恭喜你！第三次猜中了！")
        else:
            print("三次机会用完了！你没有猜中。")
```

## while 循环语句

```python
# while 条件：
#     条件满足时，要做的事情1
#     条件满足时，要做的事情2
#     条件满足时，要做的事情3
#     ......

# 循环一般需要设置终止条件，否则将无限循环
```

```python
# 运用 while 循环，计算从 1 累加到 100 的和
i = 0
result = 0
while i < 100:
    i += 1
    result += i
print(f"从 1 累加到 100 的和为:{result}")
```

```python
# 猜数字：判断输入的数字是否等于随机数

import random
num = random.randint(1,100)

count = 0

# 通过一个布尔类型的变量，做循环是否继续的标记
flag = True

while flag:
    guess_num = int(input("请猜一个 1-100 的数字:"))
    count += 1
    if guess_num == num:
        print("猜中啦！")
        # 循环终止条件
        flag = False
    else:
        if guess_num > num:
            print("你猜大了!请重新猜。")
        else:
            print("你猜小了!请重新猜。")
    
print(f"你猜对啦！答案是：{num}, 你猜了{count}次！")
```

## continue 和 break

```python
# continue 中断本次循环，直接进入下一次循环。在 for 和 while 中效果一致。
for i in range(3):
    print("语句1")
    continue
    print("语句2")
```

```python
# continue 只能控制：它所在的循环临时中断
for i in range(2):
    print("语句 1")
    
    for i in range(2):
        print("语句 2")
        continue
        print("语句 3")
    
    print("语句 4")
```

```python
# break 直接结束循环。
for i in range(3):
    print("语句1")
    break
    print("语句2")
# break后面的都不执行了，跳出循环。
```

```python
for i in range(2):
    print("语句 1")
    
    for i in range(2):
        print("语句 2")
        break
        print("语句 3")
    
    print("语句 4")
```

```python
# 案例：发工资
import random

account = 10000

for i in range(1,21):
    
    num = random.randint(1,10)
    
    if num < 5:
        print(f"员工{i},绩效分{num}，低于5，不发工资，下一位。")
        continue
        
    if account > 1000:
        account -= 1000
        print(f"向员工{i},发放工资1000元，账户余额还剩余{account}。")
    else:
        print("工资发完啦！")
        break
```

## 函数的定义

```python
# 定义一个函数 def : 已组织好的，可重复使用，针对特定功能。

str1 = "Anna"
str2 = "Honey"
str3 = "Yoyo"

def my_len(data):
    count = 0
    for i in data:
        count += 1
    print(f"字符串{data}的长度是: {count}")

my_len(str1)
my_len(str2)
my_len(str3)
```

```python
# 函数的定义：
def 函数名（传入参数）：
    函数体
    return 返回值

# 传入参数、返回值不需要，均可省略
# 函数必须先定义后使用
```

## None 类型

```python
# 如果函数没有使用 return 语句返回数据，函数实际上是有返回值的。
# 特殊的字面量：None, 其类型是：<class 'NoneType>
# 无返回值函数，实际上就是返回了：None 这个字面量，表示这个函数没有返回什么有意义的内容，即返回空。

# 用在 if 判断上，None 等同于 False。
# 一般用于在函数中主动返回 None,配合 if 判断做相关处理。

# None 用于声明无初始内容的变量
```

```python
def say_hi():
    print("你好呀！")

result = say_hi()
print(f"无返回函数，返回的内容是：{result}")
print(f"无返回函数，返回的内容类型是：{type(result)}")
```

```python
# 主动返回 None 的函数
def say_hi():
    print("你好呀！")
    return None
    
result = say_hi()
print(f"无返回函数，返回的内容是：{result}")
print(f"无返回函数，返回的内容类型是：{type(result)}")
```

```python
# None 用于 if 判断
def check_age(age):
    if age > 18:
        return 'Success'
    else:
        return None

result = check_age(16)
if not result:
    # 进入 if 表示 result 是 None 值，也就是 False
    print("未成年，不可进入。")
```

### while 循环  

```python
index = 0
while index < len(列表):
    元素 = 列表[index]
    对元素进行处理
    index += 1
```

```python
def list_while_func():
    my_list = ["A", "B", "C"]

    # 定义一个变量用来标记列表的下表
    index = 0  # 初始值为 0

    while index < len(my_list):
        element = my_list[index]
        
        print(f"列表的元素：{element}")
        
        index += 1

list_while_func()
```

### for 循环

```python
def list_for_func():

    my_list = [1, 2, 3, 4, 5]
    for element in my_list:
        print(f"列表的元素：{element}")

list_for_func()
```

### 常用操作 -- 切片

```python
# 语法：序列[起始下标：结束下标:步长]
# 结束下标：不包含，结束下标 留空 表示结尾，得到一个新序列
# 步长：默认是1，有时可省略不写
```

```python
my_list = [0,1,2,3,4,5,6]

# 步长默认为 1，从 1开 始，4 结束
result1 = my_list[1:4]

print(f"结果1：{result1}")
# 从头开始，到尾结束，步长为 1
result2 = my_list[:]
print(f"结果2：{result2}")

# 从头开始，到尾结束，步长为 2
result3 = my_list[::2]
print(f"结果2：{result3}")

# 从头开始，到尾结束，步长为 -1
result4 = my_list[::-1]
print(f"结果2：{result4}")
```

## 函数的多种参数使用形式

### 位置参数

```python
# 调用函数时根据函数定义的参数位置来传递参数
def user_info(name,age,gender):
    print(f"您的姓名是:{name},年龄是:{age},性别是:{gender}")

user_info("Tom",18,"男")
```

### 关键字参数

```python
# 调用时通过" 键 = 值 " 形式传递参数
# 作用：让函数更加清晰、容易使用，同时也消除了参数的顺序需求
# 注意：函数调用时，如果有位置参数，位置参数必须在前，且匹配参数顺序 ，关键词参数之间不存在先后顺序
```

```python
def user_info(name,age,gender):
    print(f"您的姓名是:{name},年龄是:{age},性别是:{gender}")

user_info(name="Tom",age=18,gender="男")
user_info("Tom",age=18,gender="男")
user_info("Tom",gender="男",age=18)
```

### 缺省参数

```python
# 也称默认参数，用于定义函数，为参数提供默认值，调用函数时可不传默认参数的值
# 注意：所有位置参数必须出现在默认参数前，包括函数定义与调用
# 作用：当调用函数时，没有传递参数，默认使用缺省参数对应的值
```

```python
# 设置默认值必须统一都在最后

def user_info(name,age,gender='男'):
    print(f"您的姓名是:{name},年龄是:{age},性别是:{gender}")

user_info(name="Tom",age=17)
user_info("Anna",age=18,gender="女")
```

### 不定长参数

#### 位置传递

```python
# 注意：传递的所有参数都会被 * 后面的 args变量收集
# 它会根据传进参数的位置合并为一个元组(tuple),args是元组类型，即位置传递
```

```python
def user_info(*args):
    print(f"args 参数的类型是：{type(args)},内容是：{args}")
user_info(1, 2, 3, 'Tom', '男')
```

#### 关键字传递

```python
# 注意：参数是"键 = 值" 形式的形况下，所有"键 = 值" 都会被 ** 后面的 kwargs接收
# 同时会根据"键 = 值" 组成字典
```

```python
def user_info(**kwargs):
    print(f"args 参数的类型是：{type(kwargs)},内容是：{kwargs}")
user_info(num1 = 1, num2 = 2, num3 = 3, name = 'Tom', gender ='男')
```
